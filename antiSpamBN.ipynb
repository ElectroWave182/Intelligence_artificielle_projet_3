# Antispam à l'aide d'un classificateur naïf bayésien

#### Le but de ce projet est de réaliser un  anti-spam à l'aide d'un classificateur naïf bayésien de type bernoulli.


```python
import sklearn
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

### On importe et on prépare les données

```python
# Les données du fichier csv Spam
data=pd.read_csv('spam.csv')  
```

```python
data = data.drop ([
    'Unnamed: 0',
    'label'
], axis = 1)
data.head()
```

```python
labels = list (data['label_num'])
print ("Nb de hams :", labels.count (0))
print ("Nb de mails :", len (labels))

proportion = float (labels.count (0)) / float (len (labels))
print (proportion)
```

```python
leDico = dict ()

for identifiant in range (len (labels)):

    texte = data['text'][identifiant]
    typeMail = data['label_num'][identifiant]

    for word in texte.split ():
        try:
            leDico[word][typeMail] += 1
        except:
            leDico[word] = [0, 0]
            leDico[word][typeMail] += 1
    
print (leDico)
```
